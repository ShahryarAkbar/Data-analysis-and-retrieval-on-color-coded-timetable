## Question 02
Remove Un necessary Top Rows using pandas.

```python
import pandas as pd
file_path = 'TimeTable, FSC, Fall-2023.xlsx'
df = pd.read_excel(file_path, engine='openpyxl', header=None, skiprows=4)  
df.columns = df.iloc[0]
df.reset_index(drop=True, inplace=True)
df.set_index('Room', inplace=True)
df=df.dropna(axis=1, how='all')
df
```

## Question 03
Return all the classes of the particular subject (Ask from User) along with Time, Day and
Room Number.

```python
subject = input("Enter the subject you want to find: ")
subject_df = df.applymap(lambda x: subject in str(x))
class_number, class_time = None, None

for col in subject_df.columns:
    for idx, value in subject_df[col].items():
        if value:
            class_number = idx
            class_time = col
            break
    if class_number is not None:
        break

if class_number is not None:
    print(f"The class for '{subject}' is in room {class_number} at {class_time}.")
else:
    print(f"'{subject}' not found in the schedule.")
```

## Question 04
Read the timetable along with coloring information and display all classes of a particular
batch (User Input) in a week along with Time, Room# and Day. In short, make a custom
timetable for the section. E.g. I want to extract timetable for only AI 2022 batch. So, your
output should produce an excel file, which only contains AI 2022 batch timetable, rest of
slots are free.

```python
from openpyxl import load_workbook
from openpyxl.styles import PatternFill
from collections import defaultdict

file_path = 'TimeTable, FSC, Fall-2023.xlsx'  
workbook = load_workbook(file_path, read_only=True, data_only=True)
batch_to_view = input("Enter the batch you want to view: ")

batch_colors = defaultdict(str)

def extract_batches_and_colors(sheet):
    for row in sheet.iter_rows(min_row=1, max_row=4):
        for cell in row:
            if isinstance(cell.fill, PatternFill):
                batch_name = cell.value
                batch_color = cell.fill.fgColor.rgb
                batch_colors[batch_name] = batch_color

for sheet_name in workbook.sheetnames:
    sheet = workbook[sheet_name]
    extract_batches_and_colors(sheet)

if batch_to_view not in batch_colors:
    print(f"The batch '{batch_to_view}' was not found in the timetable.")
else:
    def extract_timetable_for_batch(sheet, day):
        data = []
        for row in sheet.iter_rows(min_row=5):
            for cell in row:
                cell_value = cell.value
                if cell_value:
                    cell_color = cell.fill.fgColor.rgb if isinstance(cell.fill, PatternFill) else None
                    if cell_color == batch_colors[batch_to_view]:
                        course = cell_value
                        room = sheet.cell(row=cell.row, column=1).value
                        time = sheet.cell(row=5, column=cell.column).value
                        data.append([day, time, room, course])

        return data

    custom_timetable = pd.DataFrame(columns=['Day', 'Time', 'Room', 'Course'])

    for sheet_name in workbook.sheetnames:
        sheet = workbook[sheet_name]
        data = extract_timetable_for_batch(sheet, sheet_name)
        if data:
            custom_timetable = custom_timetable.append(data, ignore_index=True)

    output_file_path = f'{batch_to_view}_timetable.xlsx'
    custom_timetable.to_excel(output_file_path, index=False)
    print(f"The timetable for '{batch_to_view}' has been saved to {output_file_path}")
```

## Question 05

Read the Course allocators list and map it in on the time table such that giving the name
of a teacher (User Input) all the classes assigned to him/her should be returned. Likewise,
in Part 5, make a timetable for particular teacher.

```python
teacher_name = input("Enter the name of the teacher: ")
courses_file_path = 'List of Courses at FAST School of Computing, Fall-2023.xlsx' 
courses_df = pd.read_excel(courses_file_path, header=2)  

teacher_courses_full_names = courses_df[courses_df['Course Instructor'] == teacher_name]['Course']

print(f"Courses for '{teacher_name}':")
for course in teacher_courses_full_names:
    print(course)
chosen_course_abbreviation = input("Enter the course abbreviation (initial characters): ")
timetable_file_path = 'TimeTable, FSC, Fall-2023.xlsx'  
timetable_workbook = load_workbook(timetable_file_path, read_only=True, data_only=True)
timetable_found = False
for sheet_name in timetable_workbook.sheetnames:
    sheet = timetable_workbook[sheet_name]
    teacher_timetable = pd.DataFrame(columns=['Day', 'Time', 'Room', 'Course'])
    
    for row in sheet.iter_rows(min_row=5):
        for cell in row:
            cell_value = cell.value
            if cell_value and cell_value.startswith(chosen_course_abbreviation):
                room = sheet.cell(row=cell.row, column=1).value
                time = sheet.cell(row=5, column=cell.column).value
                teacher_timetable = teacher_timetable.append({'Day': sheet_name,
                                                              'Time': time,
                                                              'Room': room,
                                                              'Course': cell_value},
                                                             ignore_index=True)
    
    if not teacher_timetable.empty:
        print(f"Timetable for '{teacher_name}' on '{sheet_name}':")
        print(teacher_timetable)
        timetable_found = True

if not timetable_found:
    print(f"No timetable data found for '{teacher_name}' with course abbreviation starting with '{chosen_course_abbreviation}'.")

```
